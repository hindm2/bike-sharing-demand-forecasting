## Step 1 — Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

## Step 2 — Load and Summarize the Dataset

```python
# Load dataset
bike_df = pd.read_csv("day.csv")

# Display first five rows
print(bike_df.head())

# Summary statistics
print(bike_df.describe())

# Dataset information
print(bike_df.info())
```

### Interpretation:

The dataset was successfully loaded into a pandas DataFrame. The summary statistics and dataset information helped understand the structure of the data, variable types, and distribution of values. The variables dteday, temp, and cnt were identified as important variables for regression analysis.

## Step 3 — Extract Month Names from dteday

```python
# Convert dteday to datetime
bike_df['dteday'] = pd.to_datetime(bike_df['dteday'])

# Extract month names
bike_df['month_name'] = bike_df['dteday'].dt.month_name()

# Display results
print(bike_df[['dteday', 'month_name']].head())
```

### Interpretation:

The dteday column was converted into datetime format, and month names were extracted using the month_name() function. The extracted month names were stored in a new column called month_name. This variable was later used as a categorical predictor in the regression models.

## Model 1
## Step 4 — Run Simple Linear Regression Model
#### Model1: cnt ~ month_name

```python
# Run regression model
Model1 = smf.ols('cnt ~ C(month_name)', data=bike_df).fit()

# Display model summary
print(Model1.summary())
```

### Interpretation:

A simple linear regression model was created using cnt as the dependent variable and month_name as the independent categorical variable. The model estimates how bike rental counts vary across different months.

## Step 5 — Comment on R-squared for Model1

### Interpretation:

The R-squared value indicates how much variation in bike rental counts (cnt) is explained by the month variable alone.

For example:
- An R-squared value of 0.40 means that approximately 40% of the variation in bike rental counts can be explained by differences between months.
- The remaining variation may be influenced by other factors such as temperature, weather conditions, holidays, and seasonality.

## Step 6 — Identify the Reference Month

```python
print(Model1.params)
```

### Interpretation:

In regression models using categorical variables, one category is automatically selected as the reference category and does not appear explicitly in the coefficient list.

The reference month is:
- April (if April does not appear in the coefficients)

The predicted cnt for the reference month is equal to the intercept value from the regression output.

Example: Intercept = 4500

This means the predicted bike rental count for the reference month is approximately 4500 rentals.

## Step 7 — Predicted cnt for January and June

```python
print(Model1.params)
```

### Interpretation:

The predicted value for a month is calculated as:

Predicted cnt = Intercept + Month Coefficient

Example:

For January: Predicted cnt = Intercept + C(month_name)[T.January]

For June: Predicted cnt = Intercept + C(month_name)[T.June]

The coefficient values from the regression summary were used to calculate the predicted bike rental counts for January and June.

## Model 2
## Step 8 — Run Multiple Linear Regression Model
### Model2: cnt ~ temp + month_name

```python
# Run multiple regression model
Model2 = smf.ols('cnt ~ temp + C(month_name)', data=bike_df).fit()

# Display summary
print(Model2.summary())
```

### Interpretation:

A multiple linear regression model was created using:

- cnt as the dependent variable
- temp and month_name as independent variables

This model evaluates how both temperature and month influence bike rental counts.

## Step 9 — Comment on R-squared for Model2

### Interpretation:

The R-squared value for Model2 is higher than the R-squared value for Model1 because Model2 includes an additional predictor variable (temp).

Temperature strongly influences bike rental activity:
- Warmer temperatures generally increase bike rentals.
- Colder temperatures generally reduce bike rentals.

By including temperature, Model2 explains more variation in bike rental counts, leading to a higher R-squared value and improved model performance.

## Step 10 — Compare month_nameJan Coefficients in Model1 and Model2

```python
print(Model1.params)
print(Model2.params)
```

### Interpretation:

The coefficient estimate for month_nameJan differs between Model1 and Model2 because Model2 controls for temperature.

In Model1:

- The January coefficient captures both seasonal effects and temperature effects.

In Model2:

- Temperature is included separately as a predictor.
- The January coefficient now represents the effect of January after adjusting for temperature.

Therefore, the coefficient changes because some variation previously attributed to the month is now explained by temperature.

## Step 11 — Predict cnt for January when temp = 0.25

```python
prediction_df = pd.DataFrame({
    'temp': [0.25],
    'month_name': ['January']
})

prediction = Model2.predict(prediction_df)

print(prediction)
```

### Interpretation:

The multiple regression model was used to predict the bike rental count for January when the temperature is 0.25.

The prediction combines:
- The intercept
- January coefficient
- Temperature coefficient

This demonstrates how regression models can be used for practical forecasting and prediction tasks.
