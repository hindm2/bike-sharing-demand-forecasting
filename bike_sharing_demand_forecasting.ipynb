```python
import pandas as pd
```

```python
df = pd.read_csv("day.csv")
```

```python
import statsmodels.formula.api as smf
```

```python
print("First 5 rows:")
display(df.head())
```

```python
print("\nDataset Information:")
df.info()
```

```python
print("\nSummary Statistics:")
display(df.describe())
```

```python
df['dteday'] = pd.to_datetime(df['dteday'])
```

```python
df['month_name'] = df['dteday'].dt.month_name()
```

```python
df[['dteday', 'month_name']].head()
```

```python
Model1 = smf.ols('cnt ~ month_name', data=df).fit()

```

```python
print(Model1.summary())
```

R_squared is 0.391 means that the month explains 40% of the variation.

April is the reference month

The reference month' predicted cnt is almost 4485

Predicted cnt for January 4485 + 1287 = 5772


Predicted cnt for June    4485 + (-2309) = 2176

```python
Model2 = smf.ols('cnt ~ temp + month_name', data=df).fit()
```

```python
print(Model2.summary())
```

The R-squared value for Model2 is 0.447


Adding the predictor temp to the model provides more information, and explains more the variation in cnt, since the temperature is an important factor that affects the bike rentals.

The coefficient estimate for January model1 is -2309


The coefficient estimate for January model2 is -852


Model2 includes both month_name and temperature (temp) as predictors. Since temperature is strongly related to bike rentals and January is typically a colder month, some of the variation previously attributed to January is now explained by temperature. As a result, the January coefficient becomes smaller in magnitude, changing from -2309 to -852.

This indicates that part of the lower bike rental count observed in January is due to colder temperatures rather than the month itself.

# Key Insights

1. Temperature is a major driver of bike rental demand. Adding temperature to the model improved its explanatory power.

2. The final model explains approximately 44.7% of the variation in bike rental demand (R² = 0.447).

3. January initially appeared to strongly reduce bike demand, but much of this effect was explained by colder temperatures.

4. Seasonal and weather-related factors play an important role in customer behavior.

5. Combining multiple predictors provides better demand forecasts than using a single variable alone.

# Business Recommendations

- Increase bike availability during warmer periods when demand is expected to rise.

- Use weather forecasts to proactively adjust bike distribution across stations.

- Schedule maintenance activities during colder months when demand is lower.
